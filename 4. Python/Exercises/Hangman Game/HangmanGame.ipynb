```python
import random
import streamlit as st 

# list of words player might have to guess
mylist = ["mangosteen","lychee","guava","pomegranate","cactus","granadilla","pomelo","starfruit","jackfruit","acai"]

# random function to select one word from the list 
word_to_guess = random.choice(mylist).lower()

#Â intialise empty list for the player to guess
guessed_letters = []

# set the maximum number of incorrect guesses allowed = 6
tries = 6

# define a function to return the word with blanks if the letters not yet guessed, and for those guess show it
# " ".join puts a space between the letters
def display_word():
    return " ".join([letter if letter in guessed_letters else "_" for letter in word_to_guess])

# game loop
# first set the number of tries the player is allowed to have as before the loop

while tries > 0:
    print(f"\nWord: {display_word()}")
    input_guess = input("Guess a letter: ").lower()

    if not input_guess.isalpha() or len(input_guess) != 1:
        print("Please guess a single letter.")
        continue

    if input_guess in guessed_letters:
        print("You have already guessed that letter.")
        continue

    guessed_letters.append(input_guess)

    if input_guess in word_to_guess:
        print("Good guess!")
    else:
        tries -= 1
        print(f"Wrong! You have {tries} tries left.")

    
    # checks if all letters in the game have been guessed 
    if all(letter in guessed_letters for letter in word_to_guess):
        print(f"Congratulations! The word was: ", word_to_guess)
        break
else:
    print(f"Unfortunately, you lost! The word was: ", word_to_guess)


# add an extra feature like categories for the words
```
