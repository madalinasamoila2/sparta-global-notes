```python
# Fetching via API - easier for multiple fetches
import requests
import pandas as pd


# Function to fetch abilities data from the API
def fetch_abilities_data(url):
    ''' Fetches abilities data from the provided URL and returns the results'''
    try:
        response = requests.get(url)
        response.raise_for_status() # raise and exception for HTTP errors

        # return parsed JSON data
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from API: {e}")
        return None
    
# Function to extract ability ID
def extract_ability_id(url):
    ''' Extract the ability ID from the given URL'''
    try:
        ability_id =int(url.rstrip('/').split('/')[-1])
        return ability_id
    except ValueError as e:
        print(f"Error extracting ability ID from URL: {e}")
        return None
    
# Main code execution
url = 'https://pokeapi.co/api/v2/ability?limit=100&offset=0'

# Fetch abilities data
ability_data = fetch_abilities_data(url)

if ability_data:
    ability_dict_list = []
    
    # Extract ID and name for each ability and store them in a list
    for ability in ability_data.get('results', []):
        ability_id = extract_ability_id(ability['url'])
        if ability_id is not None:
            ability_dict_list.append({
                'id': ability_id,
                'name': ability['name']
            })
    
    # Convert the list to a DataFrame
    df = pd.DataFrame(ability_dict_list)
    
    # Convert DataFrame to CSV
    df.to_csv("pokemon_abilities.csv", index=True)
    print("DataFrame has been converted to 'pokemon_abilities.csv'")
else:
    print("Failed to retrieve ability data.")

```
